```python
# Execute the correlation matrix function

import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

mining_data = pd.read_csv(r'C:\Users\owner\Flotation_Data.csv')
plt.figure(figsize=(15, 12))
corrMatrix = mining_data.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
# Split the data into the independent and dependent variables

flotation_parameters = mining_data.iloc[:,2:23] 

reagents = mining_data.iloc[:, [0,1]]
starch_flow = reagents.iloc[:,0].values
amina_flow = reagents.iloc[:,1].values

print('Shape of flotation_parameters = ', flotation_parameters.shape)
print('Shape of starch_flow = ', starch_flow.shape)
print('Shape of amina_flow = ', amina_flow.shape)
```

```python
# Begin the execution of the p-values derivation

import statsmodels.api as sm
b_zero = np.append(arr = np.ones((len(mining_data.iloc[:,1].values),1)).astype(int),
                                                                     values = flotation_parameters,
                                                                     axis = 1)

starch_flow = reagents.iloc[:,0].values
amina_flow = reagents.iloc[:,1].values

model_starch = sm.OLS(endog = starch_flow,exog = b_zero).fit()
print(model_starch.summary())
```

```python
model_amina = sm.OLS(endog = amina_flow,exog = b_zero).fit()
print(model_amina.summary())
```

```python
pValue_st = model_starch.pvalues
pValue_am = model_amina.pvalues
pValue_st_list = []
pValue_am_list = []
for i in range(len(pValue_st)):
    if i > 0:
        pValue_st_list.append(pValue_st[i])
        pValue_am_list.append(pValue_am[i])
```

```python
# Plot the p-values graph

from sklearn.metrics import r2_score
from pylab import *

width_in_inches = 15
height_in_inches = 13
dots_per_inch = 60

plt.figure(
    figsize=(width_in_inches, height_in_inches),
    dpi=dots_per_inch)

t = flotation_parameters.columns
st = pValue_st_list
am = pValue_am_list
plot(t, st, label = 'st', color = 'blue',marker='o')
plot(t, am, label = 'am',color = 'red', marker='x')
plt.xticks(rotation=90)
plt.legend(loc="upper right",prop={"size":25})
ylabel('P Value',fontsize = 20)
title('P Value of starch and amina flows',fontsize = 25)
grid(True)
plt.figure(figsize=(60,3))
plt.savefig("p_values.png")
show()
```

```python
# Drop the non-essential variables

mining_data.drop(['Ore Pulp Flow', 'Flotation Column 02 Air Flow', 'Flotation Column 03 Air Flow', 'Flotation Column 06 Air Flow', 'Flotation Column 07 Air Flow', 'Flotation Column 01 Level','Flotation Column 02 Level','Flotation Column 03 Level','Flotation Column 04 Level','Flotation Column 05 Level','% Iron Concentrate','% Silica Concentrate'], axis=1, inplace=True)

mining_data.head(5)

#cancer_data.to_csv("C:\\Users\\owner\\Datasets\\cancer_data_.csv")
```

```python
flotation_parameters = mining_data.iloc[:,2:11] 

reagents = mining_data.iloc[:, [0,1]]
starch_flow = reagents.iloc[:,0].values
amina_flow = reagents.iloc[:,1].values

print('Shape of flotation_parameters = ', flotation_parameters.shape)
print('Shape of starch_flow = ', starch_flow.shape)
print('Shape of amina_flow = ', amina_flow.shape)
```

```python
# Split the data into train and test data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(flotation_parameters,
                                                    reagents,
                                                    test_size = 0.15,
                                                    random_state = 0)
```

```python
# Begin the development of the model

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor 

regressor_randForest = RandomForestRegressor(random_state = 0, n_estimators = 100)  
regressor_randForest.fit(x_train,y_train) 
y_pred_rf = regressor_randForest.predict(x_test)

regressor_randForest2 = RandomForestRegressor(random_state = 0, n_estimators = 100)  
regressor_randForest2.fit(flotation_parameters,reagents)
y_pred_rf2 = regressor_randForest.predict(flotation_parameters)
 
print('R2 Score of Random Forest Regression with Train-Test Split',r2_score(y_test,y_pred_rf))
print('R2 Score of Random Forest Regression with Whole Data',r2_score(reagents,y_pred_rf2))
```

```python
regressor_st = RandomForestRegressor(random_state = 0, n_estimators = 100)  
regressor_st.fit(flotation_parameters,starch_flow) 
y_pred_st = regressor_st.predict(flotation_parameters)

print('R2 Score of Random Forest Regression with Only starch',r2_score(starch_flow,y_pred_st))
```

```python
regressor_am = RandomForestRegressor(random_state = 0, n_estimators = 100)  
regressor_am.fit(flotation_parameters,amina_flow) 
y_pred_am = regressor_am.predict(flotation_parameters)

print('R2 Score of Random Forest Regression with Only amina',r2_score(amina_flow,y_pred_am))
```

```python
# Input values into the model for prediction

predictions = {'% Iron Feed':55.2,
          '% Silica Feed':16.98,
          'Ore Pulp pH':10.0697,
          'Ore Pulp Density':1.74,
          'Flotation Column 01 Air Flow':250.203,
          'Flotation Column 04 Air Flow':295.096,
          'Flotation Column 05 Air Flow':306.4,
          'Flotation Column 06 Level':443.682,
          'Flotation Column 07 Level':425.679,
          }
```

```python
# Test the model

predict_values = []
for value in predictions.values(): 
    predict_values.append(value)

predict_st = regressor_st.predict([predict_values])
predict_am = regressor_am.predict([predict_values])

print('Predicted starch flow =',predict_st)
print('Predicted amina flow =',predict_am)
```
